# Indexing

Selecting data from tensor by index. (Similar to numpy)

```python
import torch

x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape
```

```python
x[0]
```

```python
# middle bracket (dim=1)
# x[0,0] = x[0][0]

x[0][0]
```

```python
# 3rd dimension

x[0][1][2]
```

```python
x[0][1]
```

```python
# use : to select all of a target dimension

x[:, 0]
```

```python
# get all values of 0th and 1st dimension, but only index 1 of 2nd dimension

x[:, :, 1]
```

```python
x[0, :, :]
```

```python
x[0, 0, :]
```

```python
x[0, :, 2]
```
