# Reproducibility

Trying to take out random of random.

In short how a NN learns:

- start with random numbers
- tensor operations
- update random numbers


```python
import torch
```

```python
torch.rand(3, 3)
```

```python
torch.rand(3, 3) == torch.rand(3, 3)
```

```python
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
t1 = torch.randn(3, 3)

torch.manual_seed(RANDOM_SEED)
t2 = torch.randn(3, 3)

t1, t2, t1 == t2
```
