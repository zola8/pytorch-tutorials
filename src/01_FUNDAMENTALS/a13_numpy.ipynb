Numpy is a popular scientific computing library.

- Data in numpy array, want in PyTorch tensor. -> `torch.from_numpy(ndarray)`
- tensor -> numpy = `torch.Tensor.numpy()`

```python
# numpy to torch
import numpy as np
import torch

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)

array, tensor
```

```python
array.dtype
```

```python
tensor = torch.from_numpy(array).type(torch.float32)

tensor.dtype
```

```python
array, tensor
```

```python
# change value in array
array = array + 1

array, tensor
```

```python
# tensor -> numpy

tensor = torch.ones(7)
nparray = tensor.numpy()

tensor, nparray
```

```python
tensor = tensor + 1

tensor, nparray
```
