# Reshapeing, stacking, squeezing and unsqueezing tensors

- Reshape - reshape an input to a defined shape
- View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
- Stacking - combine multiple tensors on each other (vstack) or side by side (hstack)
- Squeeze - removes all '1' dimensions from a tensor
- Unsqueeze - add a '1' dimension to a target tensor
- Permute - return a view of the input with dimensions permuted (swapped) in a certain way

```python
import torch

x = torch.arange(1., 10.)
x, x.shape
```

```python
# add an extra dimension

# x_reshaped = x.reshape(1, 7)
# RuntimeError: shape '[1, 7]' is invalid for input of size 9

x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
x = torch.arange(1., 11.)
x_reshaped = x.reshape(2, 5)
x_reshaped, x_reshaped.shape
```

```python
x = torch.arange(1., 10.)
z = x.view(1, 9)
z, z.shape
# z is just a different view of x
```

```python
z[:, 0] = 5
z, x
```

```python
# stack
torch.stack([x, x, x])
```

```python
torch.stack([x, x, x], dim=1)
```

```python
# squeeze, unsqueeze
torch.squeeze(x)
```

```python
x = torch.zeros(2, 1, 2, 1, 2)
x.size()
```

```python
y = torch.squeeze(x)
y.size()
```

```python
y = torch.squeeze(x, 0)
y.size()
```

```python
y = torch.squeeze(x, 1)
y.size()
```

```python
y = torch.squeeze(x, (1, 2, 3))
y.size()
```

```python
x.shape
```

```python
y = torch.squeeze(x)
y.shape
```

```python
y.unsqueeze(0).shape
```

```python
y.unsqueeze(1).shape
```

```python
# permute
x = torch.rand(2, 3, 5)
x.shape
```

```python
torch.permute(x, (2, 0, 1)).shape
```

```python
x_original = torch.rand(size=(224, 224, 3))  # height, width, color_channel
x_original.shape
```

```python
# permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1)
# color_channel, height, width
```

```python
x = torch.zeros(5, 5, 3)
x[:, :, 2] = 1
# it is probably wrong!
x
```

```python
x_permuted = x.permute(2, 0, 1)
x_permuted
```
