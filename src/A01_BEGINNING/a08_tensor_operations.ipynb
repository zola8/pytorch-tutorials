```python
%%time
torch.matmul(tensor, tensor)
```

matrix multiplication:

- inner dimension must match
    - `(2,3) @ (2,3)` won't work
    - `(2,3) @ (3,2)` will work

- the resulting matrix has the outer dimension
    - `(2,3) @ (3,2)` -> `(2,2)`

```python
tensor @ tensor
```
