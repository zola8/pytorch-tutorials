```python
import torch

tensor = torch.tensor([1, 2, 3])
tensor
```

```python
# float64 tensor
float_tensor = torch.tensor([1.5, 2.7, 3.3], dtype=torch.float64)
float_tensor
```

```python
float_tensor.dtype
```

```python
# most important params

float_tensor = torch.tensor([1.5, 2.7, 3.3],
                            dtype=torch.float32,
                            device="cpu",
                            requires_grad=False)
float_tensor
```

- dtype - what datatype is the tensor
- device - what device is your tensor on (cpu/cuda)
- requires_grad = wheter or not to track gradients with this tensor operations
