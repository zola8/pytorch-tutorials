Let's create some *known* data using linear regression.
We use the linear regression formula to make a straight line with known parameters.

linear regression formula: Y = mX + b

```python
# known parameters
weight = 0.7
bias = 0.3

# data
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
# X = torch.arange(start, end, step)
# adds an extra dimension
y = weight * X + bias

X[:10], y[:10]
```
